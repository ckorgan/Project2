```python
import pandas as pd
```

```python
df = pd.read_csv("Enrollments Generated 02-12-19.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
datatype={"ZIP": str}
```

```python
zipcodes=pd.read_csv("https://gist.githubusercontent.com/erichurst/7882666/raw/5bdc46db47d9515269ab12ed6fb2850377fd869e/US%2520Zip%2520Codes%2520from%25202013%2520Government%2520Data",dtype=datatype)
```

```python
zipcodes.head(100)
```

```python
merge_df=pd.merge(df,zipcodes,how="left", left_on="ZIP/Postal Code", right_on="ZIP")
```

```python
merge_df.head()
```

```python
merge_df.info()
```

```python
merge_df.fillna(0,inplace = True)
```

```python
merge_df.to_csv("HistoricalEnrollments.csv")
```
